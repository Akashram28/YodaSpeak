```python
def translate(x):
    first_part = []
    second_part = []
    has_root_reached = False
    punct = "."
    if "\?" in x:
        punct = "?"
    elif "!" in x:
        punct = "!"
    x = preprocess(x)
    doc = nlp(x)
    
    for token in doc:
        if not has_root_reached:
            if len(first_part) == 0:
                first_part.append(token.text.lower() if (token.pos_ != "PROPN" and token.text != "I") else token.text)
            else:
                first_part.append(token.text)
            if token.dep_ == 'ROOT':
                has_root_reached = True
        else:
            second_part.append(token.text)
    x =  (" ".join(second_part) + ", "+ " ".join(first_part) + punct).replace(" ,,", ",")
    x = x[0].upper() + x[1:]
    return x
```

```python
print(translate("Are you going to the gym ?"))
```
